```python
import dac
import torchaudio
import torch
from IPython.display import Audio
```

```python
weights = torch.load("./models/dac_hifi_model.pth", map_location=torch.device('cpu'))
model = dac.DAC()
model.load_state_dict(weights)
model.eval()
```

```python
filename = "./samples/noisy_audio.wav"
noisy_audio, sample_rate = torchaudio.load(filename)
noisy_audio = noisy_audio.reshape(1, 1, -1)

with torch.no_grad():
    recon = model(noisy_audio)["audio"]

torchaudio.save("./samples/reconstructed.wav", recon.reshape(1,-1), sample_rate)
```
